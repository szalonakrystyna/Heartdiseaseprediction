```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from xgboost import XGBClassifier
import joblib

# VIS
sns.set(style="whitegrid")
plt.rcParams['figure.dpi'] = 120

# Paths
from google.colab import files
uploaded = files.upload()

DATA_PATH = "heart.csv"
OUT_DIR = "figures"
MODEL_DIR = "models"
os.makedirs(OUT_DIR, exist_ok=True)
os.makedirs(MODEL_DIR, exist_ok=True)

# 1. Data reading
df = pd.read_csv(DATA_PATH)
print("Data size:", df.shape)
display(df.head())

# 2. Info + missings
print("\n--- Informacje o ramce danych ---")
print(df.info())

print("\n--- Statystyki opisowe ---")
display(df.describe(include='all').T)

# Columns with missing data

df.columns
print("\n--- Liczba brakw w kolumnach ---")
missing = df.isna().sum().sort_values(ascending=False)
print(missing[missing > 0])
# If missing=0, to nic do imputacji. Jeli s, poniej przykad imputacji median:
# df['chol'] = df['chol'].fillna(df['chol'].median())

# 3. Quick checks
# Jeeli kategorie s zakodowane liczbami (np. sex 0/1), upewnijmy si jakie maj wartoci:
for col in df.select_dtypes(include='int').columns:
    print(col, "-> unique:", np.unique(df[col])[:10])

# 4. EDA + visualizations

df.columns = df.columns.str.lower()
print(df.columns)

# 4.1 Age histogram
plt.figure(figsize=(6,4))
plt.hist(df['age'].dropna(), bins=30)
plt.xlabel('Wiek')
plt.ylabel('Liczba osb')
plt.title('Histogram: Wiek pacjentw')
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "age_hist.png"))
plt.show()

# 4.2 Correlation heatmap
plt.figure(figsize=(10,8))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Macierz korelacji')
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "correlation_heatmap.png"))
plt.show()

# 4.3 Scatter: Age vs cholesterol
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='age', y='chol', hue='condition', palette='Set1', alpha=0.8)
plt.xlabel('Wiek')
plt.ylabel('Cholesterol (mg/dl)')
plt.title('Wiek vs Cholesterol (kolor = condition)')
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "age_vs_chol.png"))
plt.show()

# 4.4 Boxplot: rozkad cinienia spoczynkowego (trestbps) wg condition
plt.figure(figsize=(6,4))
sns.boxplot(x='condition', y='trestbps', data=df)
plt.xlabel('Condition (0 = brak, 1 = choroba)')
plt.ylabel('Cinienie spoczynkowe (trestbps)')
plt.title('Trestbps rozkad wg condition')
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "trestbps_boxplot.png"))
plt.show()

# 5. Pivot table (redni wiek wg pci i condition)
pivot_age = df.pivot_table(values='age', index='sex', columns='condition', aggfunc='mean')
print("\nPivot table - redni wiek wg sex x condition:")
display(pivot_age)

# 6. Przygotowanie danych do modelu
# Zakadamy, e 'condition' to etykieta (1 = choroba serca, 0 = brak)
condition = 'condition'
X = df.drop(columns=[condition])
y = df[condition].copy()

# Rozdzielamy kolumny numeryczne i kategoryczne
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()
# Jeeli dataset ma kategorie zakodowane liczbami (np. 'sex', 'cp'), moemy je traktowa jako kategorie:
# Dla pewnoci wypiszmy kolumny:
print("Numeryczne kolumny:", num_cols)
print("Kategoryczne kolumny:", cat_cols)

# Jeli nie ma kolumn typu object, moemy rcznie okreli kilka jako kategoryczne (np. 'sex','cp','thal')
manual_cat = []
for c in ['sex','cp','fbs','restecg','exang','slope','ca','thal']:
    if c in X.columns:
        manual_cat.append(c)
# Usu je z num_cols:
num_cols = [c for c in num_cols if c not in manual_cat]
cat_cols = manual_cat

print("Finalne num_cols:", num_cols)
print("Finalne cat_cols:", cat_cols)

# Pipeline preprocessingu
numeric_features = [
    'age', 'trestbps', 'chol', 'thalach', 'oldpeak'
]

categorical_features = [
    'sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'
]

numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)


# 7. Podzia na zbir treningowy/testowy
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# 8. Model XGBoost w pipeline
clf = Pipeline(steps=[
    ('preproc', preprocessor),
    ('xgb', XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42))
])

# Prosty grid search (may)  mona rozszerzy
param_grid = {
    'xgb__n_estimators': [50, 100],
    'xgb__max_depth': [3, 5],
    'xgb__learning_rate': [0.1, 0.01]
}

grid = GridSearchCV(clf, param_grid, cv=5, scoring='roc_auc', n_jobs=-1, verbose=1)
grid.fit(X_train, y_train)

print("Najlepsze parametry:", grid.best_params_)
best_model = grid.best_estimator_

# 9. Ewaluacja na zbiorze testowym
y_pred = best_model.predict(X_test)
y_proba = best_model.predict_proba(X_test)[:,1]

print("\n--- Classification report (test) ---")
print(classification_report(y_test, y_pred))

roc_auc = roc_auc_score(y_test, y_proba)
print("ROC AUC (test):", roc_auc)

# Rysujemy ROC curve
fpr, tpr, _ = roc_curve(y_test, y_proba)
plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, label=f'ROC (AUC = {roc_auc:.3f})')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - XGBoost')
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "roc_curve.png"))
plt.show()

# 10. Feature importance (po preprocessie; musimy zmapowa nazwy cech)
# Aby zmapowa feature names po OneHotEncoder:
preproc = best_model.named_steps['preproc']
# Zawiera transformacje: num_cols oraz ohe.get_feature_names_out(cat_cols)
num_names = num_cols
cat_names = []
if cat_cols:
    ohe = preproc.named_transformers_['cat'].named_steps['ohe']
    cat_names = list(ohe.get_feature_names_out(cat_cols))
feature_names = num_names + cat_names

xgb_model = best_model.named_steps['xgb']
importances = xgb_model.feature_importances_

feat_imp = pd.DataFrame({'feature': feature_names, 'importance': importances})
feat_imp = feat_imp.sort_values(by='importance', ascending=False).head(25)
plt.figure(figsize=(8,6))
sns.barplot(x='importance', y='feature', data=feat_imp)
plt.title('Top feature importances (XGBoost)')
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "feature_importance.png"))
plt.show()

# 11. Zapis modelu i wynikw
joblib.dump(best_model, os.path.join(MODEL_DIR, "xgb_heart_best.joblib"))
print("Model zapisany w:", os.path.join(MODEL_DIR, "xgb_heart_best.joblib"))

# Zapis podstawowych metryk do pliku
with open(os.path.join(MODEL_DIR, "metrics.txt"), "w") as f:
    f.write(f"ROC_AUC_test: {roc_auc}\n")
    f.write(classification_report(y_test, y_pred))

print("Wszystkie wykresy zapisane w katalogu:", OUT_DIR)


import os
import matplotlib.pyplot as plt

# 1. Utworzenie folderu 'figures' jeli nie istnieje
fig_dir = "figures"
os.makedirs(fig_dir, exist_ok=True)

# 2. Funkcja zapisujca wykres na podstawie tytuu
def save_fig_from_title():
    title = plt.gca().get_title()     # pobiera tytu obecnego wykresu
    if title.strip() == "":
        raise ValueError("Wykres nie ma tytuu  nie mona zapisa!")

    filename = title.replace(" ", "_").replace("/", "_") + ".png"
    filepath = os.path.join(fig_dir, filename)
    plt.savefig(filepath, dpi=300, bbox_inches='tight')
    print(f" Zapisano: {filepath}")

# Koniec
```
